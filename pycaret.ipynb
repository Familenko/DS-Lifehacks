# pycaret

```python
import pandas as pd
from pycaret.classification import *
from sklearn.datasets import load_iris

# Завантаження даних
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

```python
# Налаштування середовища PyCaret для класифікації
clf_setup = setup(data=df, target='target', session_id=123)
```

```python
models_info = models()
models_info
```

```python
# Порівняння моделей
best_model = compare_models()
```

```python
# Створення моделі
model = create_model('xgboost')
```

```python
params = model.get_params()
```

```python
params
```

```python
# Налаштування моделі
tuned_model = tune_model(model, search_library='scikit-optimize', search_algorithm='bayesian', n_iter=20, optimize='Recall')
```

```python
tuned_model
```

```python
# Оцінка моделі
evaluate_model(tuned_model)
```

```python
# Збереження моделі
save_model(tuned_model, 'data/best_model')
```
