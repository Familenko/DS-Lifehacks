```python
import pandas as pd 
import seaborn as sns 
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv('data/Google_Stock_Train (2010-2022).xlsx - in.csv')
```

```python
df
```

```python
sns.set_theme() 
sns.lineplot(df, x='Date', y='Open')
```

```python
results = seasonal_decompose(df['Open'], period=365)
```

```python
results.seasonal.plot(title='Seasonal Component', figsize=(16,4), color='blue', ylabel='Seasonality', xlabel='Date')
```

```python
results.trend.plot(title='Trend Component', figsize=(16,4), color='red', ylabel='Trend', xlabel='Date')
```

```python
results.resid.plot(title='Residual Component', figsize=(16,4), color='green', ylabel='Residuals', xlabel='Date')
```

```python
figure = results.plot() 

figure.set_figwidth(16) 
figure.set_figheight(6)
```
